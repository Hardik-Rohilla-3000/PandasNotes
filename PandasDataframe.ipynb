How to create a dataframe

```python
#How to create a dataframe
import pandas as pd
import numpy as np
np.random.seed(101)
mydata = np.random.randint(0,101,(4,3))
mydata
```

```python
myindex = ["CA","NY","AZ","TX"]
mycolumns = ["Jan","Feb","Mar"]
df = pd.DataFrame(mydata)
df
```

```python
df = pd.DataFrame(data=mydata,index=myindex)
df
```

```python
df = pd.DataFrame(data=mydata,index=myindex,columns=mycolumns)
df
```

```python
df.info()
```

Working with columns

```python
df['total_bill']
```

```python
df[['total_bill','tip']]
```

```python
df['tip'] + df['total_bill']
```

```python
df['tip_percentage'] = (df['tip']/df['total_bill'])*100   #how to add a new column
df.head()
```

```python
df['price_per_person'] = np.round(df['total_bill']/df['size'],2) #np.round(data we want to roundoff,no of digits after decimal)
df.head()
```

```python
df.drop('tip_percentage',axis=1)  #how to delete a column axis=0 (rows) axis=1(columns) by default axis=0
                                  #this change is not permanent
```

```python
df = df.drop('tip_percentage',axis=1)  #this will be permanent (recommended)
                                       #alternative df.drop('tip_percentage',axis=1,inplace=True)
```
