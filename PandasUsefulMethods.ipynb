Apply method to use custom python functions on single column

```python
df = pd.read_csv("tips.csv")
df.head()
```

```python
def last_four(num):
    return str(num)[-4:]
```

```python
df['CC Number'].apply(last_four)
```

```python
df['last_four'] = df['CC Number'].apply(last_four)
```

```python
df['total_bill'].mean()
```

```python
def yelp(price):
    if price < 10:
        return "$"
    elif price >= 10 and price < 30:
        return "$$"
    else:
        return "$$$"
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
```

Using Lambda expression for simple functions in apply method

def simple(num):
    return num*2
This function can be written in lambda expresion like this
lambda num : num*2

```python
df['total_bill'].apply(lambda num: num*2)  
```

Using apply methods in multiple columns

```python
def quality(total_bill,tip):
    if (tip/total_bill) > 25:
        return "Generous"
    else:
        return "Other"
```

```python
df["Quality"] = df[['total_bill','tip']].apply(lambda df:quality(df['total_bill'],df['tip']),axis=1)
df['Quality']
```

Using np.vectorize() method for multiple columns (faster to execute and simpler syntax)

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
```
